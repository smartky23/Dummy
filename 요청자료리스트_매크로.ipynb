```python
import pandas as pd

file_paths = [
    '/content/220203 요청자료 List_가온브로드밴드 v17시.xlsx',
    '/content/220205 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드_회사회신.xlsx'
]

dfs = {}
for file_path in file_paths:
    try:
        df_name = file_path.split('/')[-1].split('.')[0]
        # Read Excel starting from the row with actual headers (row 6, index 5)
        dfs[df_name] = pd.read_excel(file_path, header=5)
        print(f"'{file_path}' 파일 로드 성공. DataFrame 이름: '{df_name}'")
        display(dfs[df_name].head())
    except Exception as e:
        print(f"'{file_path}' 파일 로드 중 오류 발생: {e}")
```

**버전 관리 및 진행률 기록**: 각 버전별 요청 자료 목록과 수령 여부 상태를 기록하고, 전체 진행률을 계산하여 관리할 수 있는 방법을 구현합니다.

```python
# Store processed dataframes with submission status
version_history = {}

for df_name, df in dfs.items():
    # Drop rows where 'Unnamed: 2' is NaN, as these are likely not actual request items
    df_cleaned = df.dropna(subset=['Unnamed: 2']).copy() # Use .copy() to avoid SettingWithCopyWarning

    # Calculate progress for each version
    total_items = len(df_cleaned)
    received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '수령'].shape[0]
    partially_received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '일부수령'].shape[0]

    # Calculate progress percentage (considering partially received as half for simplicity)
    # You can adjust this calculation based on how you define "progress"
    progress_percentage = ((received_items + (partially_received_items * 0.5)) / total_items) * 100 if total_items > 0 else 0

    version_history[df_name] = {
        'dataframe': df_cleaned,
        'progress_percentage': progress_percentage,
        'submission_counts': df_cleaned['Unnamed: 8'].value_counts(dropna=False).to_dict()
    }

# Display the calculated progress for each version
for version, data in version_history.items():
    print(f"버전: {version}")
    print(f"  진행률: {data['progress_percentage']:.2f}%")
    print(f"  제출 여부 개수: {data['submission_counts']}")
    print("-" * 30)
```

## 데이터 로딩 및 기본 가공

### Subtask:
각 버전별 엑셀 파일을 불러오고, 필요한 컬럼을 정리합니다. (기존 로드된 데이터 활용)


**Reasoning**:
Process each dataframe in `dfs` by selecting and renaming relevant columns, dropping rows with NaN in 'Unnamed: 2', and store the results in a new dictionary `processed_dfs`.



```python
processed_dfs = {}

for df_name, df in dfs.items():
    # Select relevant columns and rename them
    if 'Unnamed: 1' in df.columns and 'Unnamed: 2' in df.columns and 'Unnamed: 3' in df.columns and 'Unnamed: 8' in df.columns:
        processed_df = df[['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 8']].copy()
        processed_df.columns = ['계정', 'No.', '요청자료List', '제출여부']

        # Drop rows where 'No.' (originally 'Unnamed: 2') is NaN
        processed_df = processed_df.dropna(subset=['No.'])

        processed_dfs[df_name] = processed_df
        print(f"'{df_name}' 데이터프레임 처리가 완료되었습니다.")
        display(processed_dfs[df_name].head())
    else:
        print(f"'{df_name}' 데이터프레임에 필요한 컬럼이 부족하여 처리를 건너뜁니다.")
```

## 버전 간 비교 - 신규 항목 파악

### Subtask:
이전 버전에는 없었지만 현재 버전에 새로 추가된 요청 자료 항목을 식별합니다.


**Reasoning**:
Iterate through the processed dataframes, comparing each version to the previous one to identify new rows based on the combination of 'No.' and '요청자료List' columns. Store the new items along with the version where they were added.



```python
new_items_history = {}
versions = list(processed_dfs.keys())

for i in range(1, len(versions)):
    current_version_name = versions[i]
    previous_version_name = versions[i-1]

    current_df = processed_dfs[current_version_name]
    previous_df = processed_dfs[previous_version_name]

    # Create a combined key for comparison
    current_combined = current_df.apply(lambda row: (row['No.'], row['요청자료List']), axis=1)
    previous_combined = previous_df.apply(lambda row: (row['No.'], row['요청자료List']), axis=1)

    # Find new items in the current version not present in the previous version
    new_items_mask = ~current_combined.isin(previous_combined)
    new_items_df = current_df[new_items_mask].copy()

    if not new_items_df.empty:
        new_items_history[current_version_name] = new_items_df

# Print the identified new items
print("새로 추가된 요청 자료 항목:")
for version, items_df in new_items_history.items():
    print(f"\n--- '{version}' 버전에서 새로 추가된 항목 ---")
    display(items_df[['No.', '요청자료List', '제출여부']])
```

## 버전 간 비교 - 변경된 항목 파악

### Subtask:
이전 버전과 현재 버전 간에 제출 여부 등 내용이 변경된 요청 자료 항목을 식별합니다.


**Reasoning**:
Iterate through the processed dataframes, skipping the first version, and compare the current version with the previous version to identify items where the submission status has changed.



```python
changed_items_history = {}
versions = list(processed_dfs.keys())

for i in range(1, len(versions)):
    current_version_name = versions[i]
    previous_version_name = versions[i-1]

    current_df = processed_dfs[current_version_name]
    previous_df = processed_dfs[previous_version_name]

    # Merge the dataframes on 'No.' to compare items present in both versions
    merged_df = pd.merge(previous_df, current_df, on='No.', suffixes=('_prev', '_curr'), how='inner')

    # Identify rows where the '제출여부' status has changed
    changed_status_mask = merged_df['제출여부_prev'] != merged_df['제출여부_curr']
    changed_items_df = merged_df[changed_status_mask].copy()

    # Select and rename the relevant columns for the output
    if not changed_items_df.empty:
        changed_items_df = changed_items_df[['No.', '요청자료List_curr', '제출여부_prev', '제출여부_curr']]
        changed_items_df.columns = ['No.', '요청자료List', '이전 제출여부', '현재 제출여부']
        changed_items_history[current_version_name] = changed_items_df

# Print the identified changed items
print("제출 여부 등 내용이 변경된 요청 자료 항목:")
for version, items_df in changed_items_history.items():
    print(f"\n--- '{version}' 버전에서 변경된 항목 ---")
    display(items_df)
```

## 현재 상태 요약

### Subtask:
가장 최신 버전의 데이터를 기준으로 전체 요청 자료 수, 수령 완료 수, 일부 수령 수, 미수령 수 등을 집계합니다.


**Reasoning**:
To fulfill the subtask, I need to extract the last processed dataframe from the `processed_dfs` dictionary and then calculate the total number of requested items, the number of received items, partially received items, not received items, and items with NaN status based on the '제출여부' column.



```python
# Get the name of the last processed dataframe
last_version_name = list(processed_dfs.keys())[-1]
last_processed_df = processed_dfs[last_version_name]

# Calculate the total number of requested items
total_requested = len(last_processed_df)

# Calculate counts for each submission status, including NaN
submission_counts = last_processed_df['제출여부'].value_counts(dropna=False).to_dict()

# Extract counts for specific statuses, defaulting to 0 if a status is not present
received_count = submission_counts.get('수령', 0)
partially_received_count = submission_counts.get('일부수령', 0)
not_received_count = submission_counts.get('미수령', 0)
nan_count = submission_counts.get(pd.NA, 0) # Handle potential pandas NA

# Store the aggregated results
current_status_summary = {
    'Total Requested': total_requested,
    'Received': received_count,
    'Partially Received': partially_received_count,
    'Not Received': not_received_count,
    'NaN Status': nan_count
}

# Print the summary
print(f"--- '{last_version_name}' 버전 현재 상태 요약 ---")
print(f"전체 요청 자료 수: {current_status_summary['Total Requested']}")
print(f"수령 완료 수: {current_status_summary['Received']}")
print(f"일부 수령 수: {current_status_summary['Partially Received']}")
print(f"미수령 수: {current_status_summary['Not Received']}")
print(f"NaN 상태 수: {current_status_summary['NaN Status']}")
print("-" * 30)
```

```python
import pandas as pd
import numpy as np

# Define file_paths and load data into dfs
file_paths = [
    '/content/220203 요청자료 List_가온브로드밴드 v17시.xlsx',
    '/content/220205 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드_회사회신.xlsx'
]

dfs = {}
for file_path in file_paths:
    try:
        df_name = file_path.split('/')[-1].split('.')[0]
        # Read Excel starting from the row with actual headers (row 6, index 5)
        dfs[df_name] = pd.read_excel(file_path, header=5)
        print(f"'{file_path}' 파일 로드 성공. DataFrame 이름: '{df_name}'")
    except Exception as e:
        print(f"'{file_path}' 파일 로드 중 오류 발생: {e}")


# Process dataframes
processed_dfs = {}
for df_name, df in dfs.items():
    # Select relevant columns and rename them
    if 'Unnamed: 1' in df.columns and 'Unnamed: 2' in df.columns and 'Unnamed: 3' in df.columns and 'Unnamed: 8' in df.columns:
        processed_df = df[['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 8']].copy()
        processed_df.columns = ['계정', 'No.', '요청자료List', '제출여부']

        # Drop rows where 'No.' (originally 'Unnamed: 2') is NaN
        processed_df = processed_df.dropna(subset=['No.'])

        processed_dfs[df_name] = processed_df
        print(f"'{df_name}' 데이터프레임 처리가 완료되었습니다.")
    else:
        print(f"'{df_name}' 데이터프레임에 필요한 컬럼이 부족하여 처리를 건너뜁니다.")

# Calculate version history and progress
version_history = {}
for df_name, df in dfs.items():
    df_cleaned = df.dropna(subset=['Unnamed: 2']).copy()
    total_items = len(df_cleaned)
    received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '수령'].shape[0]
    partially_received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '일부수령'].shape[0]
    progress_percentage = ((received_items + (partially_received_items * 0.5)) / total_items) * 100 if total_items > 0 else 0

    version_history[df_name] = {
        'dataframe': df_cleaned,
        'progress_percentage': progress_percentage,
        'submission_counts': df_cleaned['Unnamed: 8'].value_counts(dropna=False).to_dict()
    }


# Calculate all_changes_history (new and changed items)
all_changes_history = {}
versions = list(processed_dfs.keys())

for i in range(1, len(versions)):
    current_version_name = versions[i]
    previous_version_name = versions[i-1]

    current_df = processed_dfs[current_version_name].copy()
    previous_df = processed_dfs[previous_version_name].copy()

    merged_df = pd.merge(previous_df, current_df, on='No.', how='outer', suffixes=('_prev', '_curr'))

    # Identify new items
    new_items_mask = merged_df['요청자료List_prev'].isna() & merged_df['요청자료List_curr'].notna()
    new_items_df = merged_df[new_items_mask].copy()
    if not new_items_df.empty:
        new_items_df['ChangeType'] = '신규 추가'
        new_items_df['변경 전 제출여부'] = ''
        new_items_df['변경 후 제출여부'] = new_items_df['제출여부_curr']
        new_items_df['요청자료List'] = new_items_df['요청자료List_curr']
        original_indices = current_df[current_df['No.'].isin(new_items_df['No.'])].index + 6
        new_items_df['셀위치'] = 'I' + original_indices.astype(str)
        new_items_df = new_items_df[['No.', '셀위치', '요청자료List', 'ChangeType', '변경 전 제출여부', '변경 후 제출여부']]


    # Identify changed items
    changed_items_mask = (merged_df['제출여부_prev'].notna()) & \
                         (merged_df['제출여부_curr'].notna()) & \
                         (merged_df['제출여부_prev'] != merged_df['제출여부_curr'])

    changed_items_df = merged_df[changed_items_mask].copy()
    if not changed_items_df.empty:
        changed_items_df['ChangeType'] = '상태 변경'
        changed_items_df['변경 전 제출여부'] = changed_items_df['제출여부_prev']
        changed_items_df['변경 후 제출여부'] = changed_items_df['제출여부_curr']
        changed_items_df['요청자료List'] = changed_items_df['요청자료List_curr']
        original_indices = current_df[current_df['No.'].isin(changed_items_df['No.'])].index + 6
        changed_items_df['셀위치'] = 'I' + original_indices.astype(str)
        changed_items_df = changed_items_df[['No.', '셀위치', '요청자료List', 'ChangeType', '변경 전 제출여부', '변경 후 제출여부']]

    # Combine new and changed items
    combined_changes_df = pd.concat([new_items_df, changed_items_df], ignore_index=True)

    if not combined_changes_df.empty:
        combined_changes_df.insert(0, '연번', range(1, 1 + len(combined_changes_df)))
        all_changes_history[f"{previous_version_name} -> {current_version_name}"] = combined_changes_df

# Calculate current status summary for the latest version
last_version_name = list(processed_dfs.keys())[-1]
last_processed_df = processed_dfs[last_version_name]

total_requested = len(last_processed_df)
submission_counts_latest = last_processed_df['제출여부'].value_counts(dropna=False).to_dict()

received_count = submission_counts_latest.get('수령', 0)
partially_received_count = submission_counts_latest.get('일부수령', 0)
not_received_count = submission_counts_latest.get('미수령', 0)
nan_count = submission_counts_latest.get(np.nan, 0) # Use np.nan for checking NaN in dictionary keys

current_status_summary = {
    'Total Requested': total_requested,
    'Received': received_count,
    'Partially Received': partially_received_count,
    'Not Received': not_received_count,
    'NaN Status': nan_count
}

print("\n--- 보고서 생성 데이터 준비 완료 ---")
```

```python
import pandas as pd
import numpy as np

# Define file_paths and load data into dfs if they are not already defined
# Assuming file_paths and dfs are available from previous successful runs
# If not, uncomment and run the following lines:
file_paths = [
    '/content/220203 요청자료 List_가온브로드밴드 v17시.xlsx',
    '/content/220205 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드.xlsx',
    '/content/220208 요청자료 List_가온브로드밴드_회사회신.xlsx'
]

dfs = {}
for file_path in file_paths:
    try:
        df_name = file_path.split('/')[-1].split('.')[0]
        # Read Excel starting from the row with actual headers (row 6, index 5)
        dfs[df_name] = pd.read_excel(file_path, header=5)
        print(f"'{file_path}' 파일 로드 성공. DataFrame 이름: '{df_name}'")
    except Exception as e:
        print(f"'{file_path}' 파일 로드 중 오류 발생: {e}")


# Process dataframes
processed_dfs = {}
for df_name, df in dfs.items():
    # Select relevant columns and rename them
    if 'Unnamed: 1' in df.columns and 'Unnamed: 2' in df.columns and 'Unnamed: 3' in df.columns and 'Unnamed: 8' in df.columns:
        processed_df = df[['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 8']].copy()
        processed_df.columns = ['계정', 'No.', '요청자료List', '제출여부']

        # Drop rows where 'No.' (originally 'Unnamed: 2') is NaN
        processed_df = processed_df.dropna(subset=['No.'])

        processed_dfs[df_name] = processed_df
        print(f"'{df_name}' 데이터프레임 처리가 완료되었습니다.")
    else:
        print(f"'{df_name}' 데이터프레임에 필요한 컬럼이 부족하여 처리를 건너뜁니다.")

# Calculate version history and progress
version_history = {}
for df_name, df in dfs.items():
    df_cleaned = df.dropna(subset=['Unnamed: 2']).copy()
    total_items = len(df_cleaned)
    received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '수령'].shape[0]
    partially_received_items = df_cleaned[df_cleaned['Unnamed: 8'] == '일부수령'].shape[0]
    progress_percentage = ((received_items + (partially_received_items * 0.5)) / total_items) * 100 if total_items > 0 else 0

    version_history[df_name] = {
        'dataframe': df_cleaned,
        'progress_percentage': progress_percentage,
        'submission_counts': df_cleaned['Unnamed: 8'].value_counts(dropna=False).to_dict()
    }


# Calculate all_changes_history (new and changed items)
all_changes_history = {}
versions = list(processed_dfs.keys())

for i in range(1, len(versions)):
    current_version_name = versions[i]
    previous_version_name = versions[i-1]

    current_df = processed_dfs[current_version_name].copy()
    previous_df = processed_dfs[previous_version_name].copy()

    merged_df = pd.merge(previous_df, current_df, on='No.', how='outer', suffixes=('_prev', '_curr'))

    # Identify new items
    new_items_mask = merged_df['요청자료List_prev'].isna() & merged_df['요청자료List_curr'].notna()
    new_items_df = merged_df[new_items_mask].copy()
    if not new_items_df.empty:
        new_items_df['ChangeType'] = '신규 추가'
        new_items_df['변경 전 제출여부'] = ''
        new_items_df['변경 후 제출여부'] = new_items_df['제출여부_curr']
        new_items_df['요청자료List'] = new_items_df['요청자료List_curr']
        original_indices = current_df[current_df['No.'].isin(new_items_df['No.'])].index + 6
        new_items_df['셀위치'] = 'I' + original_indices.astype(str)
        new_items_df = new_items_df[['No.', '셀위치', '요청자료List', 'ChangeType', '변경 전 제출여부', '변경 후 제출여부']]


    # Identify changed items
    changed_items_mask = (merged_df['제출여부_prev'].notna()) & \
                         (merged_df['제출여부_curr'].notna()) & \
                         (merged_df['제출여부_prev'] != merged_df['제출여부_curr'])

    changed_items_df = merged_df[changed_items_mask].copy()
    if not changed_items_df.empty:
        changed_items_df['ChangeType'] = '상태 변경'
        changed_items_df['변경 전 제출여부'] = changed_items_df['제출여부_prev']
        changed_items_df['변경 후 제출여부'] = changed_items_df['제출여부_curr']
        changed_items_df['요청자료List'] = changed_items_df['요청자료List_curr']
        original_indices = current_df[current_df['No.'].isin(changed_items_df['No.'])].index + 6
        changed_items_df['셀위치'] = 'I' + original_indices.astype(str)
        changed_items_df = changed_items_df[['No.', '셀위치', '요청자료List', 'ChangeType', '변경 전 제출여부', '변경 후 제출여부']]

    # Combine new and changed items
    combined_changes_df = pd.concat([new_items_df, changed_items_df], ignore_index=True)

    if not combined_changes_df.empty:
        combined_changes_df.insert(0, '연번', range(1, 1 + len(combined_changes_df)))
        all_changes_history[f"{previous_version_name} -> {current_version_name}"] = combined_changes_df

# Calculate current status summary for the latest version
last_version_name = list(processed_dfs.keys())[-1]
last_processed_df = processed_dfs[last_version_name]

total_requested = len(last_processed_df)
submission_counts_latest = last_processed_df['제출여부'].value_counts(dropna=False).to_dict()

received_count = submission_counts_latest.get('수령', 0)
partially_received_count = submission_counts_latest.get('일부수령', 0)
not_received_count = submission_counts_latest.get('미수령', 0)
nan_count = submission_counts_latest.get(np.nan, 0) # Use np.nan for checking NaN in dictionary keys

current_status_summary = {
    'Total Requested': total_requested,
    'Received': received_count,
    'Partially Received': partially_received_count,
    'Not Received': not_received_count,
    'NaN Status': nan_count
}

print("\n--- 보고서 생성 데이터 준비 완료 ---")
```
