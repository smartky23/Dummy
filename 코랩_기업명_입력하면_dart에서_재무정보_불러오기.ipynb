# 정기공시 재무정보 가져오기
Dart 고유번호를 통해서 연결/별도 정보를 가져옴
 - 아직 연결은 못함

```python
import requests
import pandas as pd

def get_financial_statement(corp_code: str, fs_div: str, start_year: int = 2020, end_year: int = 2024):
    """
    특정 회사의 2020~2024년 지정된 재무제표 불러오기
    :param corp_code: DART 고유번호
    :param fs_div: 재무제표 구분 (BS: 재무상태표, IS: 손익계산서, CIS: 포괄손익계산서, CF: 현금흐름표, SCE: 자본변동표)
    :param start_year: 시작 연도
    :param end_year: 종료 연도
    :return: pandas DataFrame
    """
    all_data = []

    for year in range(start_year, end_year + 1):
        # 사업보고서(11011) 기준
        params = {
            "crtfc_key": api_key,
            "corp_code": corp_code,
            "bsns_year": str(year),
            "reprt_code": "11011",  # 사업보고서
            "fs_div": fs_div # 재무제표 구분 추가
        }

        url = "https://opendart.fss.or.kr/api/fnlttSinglAcntAll.json"
        response = requests.get(url, params=params)
        data = response.json()

        if data.get("status") != "000":
            print(f"[{year}] {fs_div} 불러오기 실패: {data.get('message')}")
            continue

        df = pd.DataFrame(data.get("list", []))
        df["year"] = year
        all_data.append(df)

    if all_data:
        result = pd.concat(all_data, ignore_index=True)
        # sj_div가 인자로 받은 fs_div와 같은 항목만 선택
        fs_df = result[result["sj_div"] == fs_div]
        return fs_df

    return pd.DataFrame()  # 실패 시 빈 DataFrame 반환

# Example usage (for testing, not part of the subtask instructions)
# if 'corp_code' in locals():
#   df_is_new = get_financial_statement(corp_code, 'IS')
#   if not df_is_new.empty:
#       print(f"{corp_code} 기업의 2020년부터 2024년까지 손익계산서 데이터를 (새 함수로) 불러왔습니다.")
#       display(df_is_new.head())
#   else:
#       print(f"{corp_code} 기업의 2020년부터 2024년까지 손익계산서 데이터를 (새 함수로) 찾을 수 없습니다.")
# else:
#   print("기업의 DART 고유 번호(corp_code)가 설정되지 않았습니다.")
```


# 파일 저장

```python
import pandas as pd
import datetime

# 엑셀 파일로 저장할 피벗 테이블 딕셔너리
pivot_tables = {
   '재무상태표(BS)' : pivot_bs,
   '손익계산서(IS)' : pivot_is,
   '현금흐름표(CF)' : pivot_cf,
   '자본변동표(SCE)' : pivot_sce,
}

# 엑셀 파일 이름
excel_file_name = f'{corp_code}_{company_input}_2020~2024년 재무정보.xlsx'

# ExcelWriter 객체 생성
with pd.ExcelWriter(excel_file_name) as writer:

        for sheet_name, pivot_table in pivot_tables.items():
            # 인덱스(계정명)를 포함하여 저장
            pivot_table.to_excel(writer, sheet_name=sheet_name)

            # 열 너비 자동 조정
            worksheet = writer.sheets[sheet_name]
            for i, col in enumerate(pivot_table.columns):
                max_len = max(pivot_table[col].astype(str).str.len().max(), len(str(col))) + 2
                worksheet.set_column(i + 1, i + 1, max_len) # +1 for index column, +2 for padding

            # 인덱스 열 너비 조정
            index_max_len = max(pivot_table.index.astype(str).str.len().max(), len(pivot_table.index.name or '')) + 2
            worksheet.set_column(0, 0, index_max_len)


print(f"'{excel_file_name}' 파일로 재무제표 피벗 테이블 저장이 완료되었습니다.")
print("'Lead' 시트가 추가되었으며 각 시트의 열 너비가 자동 조정되었습니다.")
print("왼쪽 파일 탐색기에서 파일을 확인하고 다운로드하실 수 있습니다.")
```
